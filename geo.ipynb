```python
import plotly.express as px
import pandas as pd
data = pd.read_excel('final.xlsx')
# Подготовка данных
df = pd.DataFrame(data)

geojson = 'russia.geojson' 

fig = px.choropleth(df, 
                    geojson=geojson, 
                    locations='gibdd_id', 
                    featureidkey="properties.region_id",  # Путь к названиям регионов в GeoJSON
                    color='part_monogdetnye_families', #Что изображать цветом
                    color_continuous_scale="Viridis",  # Цветовая схема
                    scope="asia"
                    )  

fig.update_geos(fitbounds="locations", visible=False) 
fig.update_layout(autosize=False,
                    width=1400, height=800,
                    #margin=dict(l=65, r=50, b=65, t=90)
                    margin=dict(l=0, r=0, b=0, t=0)
                    ) 
fig.show()
```

```python
corr = pd.read_excel('onlydigits.xlsx').corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Создание тепловой карты
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Настройка отображения
plt.title('Correlation Matrix')
plt.show()
```
